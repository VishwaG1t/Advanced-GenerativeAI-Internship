# 1. User Login Check

```python
username = "admin"
password = "1234"
```

```python
# Input Credentials
InputUser = input("username :")
InputPswd = input("password :")
```

```python
# Checking credentials
if InputUser == username:
    if InputPswd == password:
        print("Login Successful")
    else:
        print("Invalid Credentials")
else:
    print("failed")
```

# 2. Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60]

# Created 2 new list to store pass and fail students
PassStud = []
FailStud = []
```

```python
# Logic to filter out passed and failed Students
for i in marks :
    if i >= 50 :
        PassStud.append(i)
    else :
        FailStud.append(i)
```

```python
# Students who passed and failed marks
print(PassStud)
print(FailStud)
```

```python
# No. of Students who passed and failed
print(len(PassStud))
print(len(FailStud))
```

# 3. Simple Data Cleaner


```python
names = [" Alice ", "bob", " CHARLIE "]

# Removeing extra spacing and converting to lower case
new_names = [x.strip().lower() for x in names]

# Printing cleaned list
print(new_names)
```

# Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]
```

```python
# length of each messages
for i in messages :
    print(i , ":" , len(i) )
```

```python
# Messages longer than 10 characters
for i in messages :
    if len(i) > 10 :
        print(i , "(flagged message)")
```

# Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counting the no. of "ERROR" entries in logs
error_count = logs.count("ERROR")

# Print the total error count
print("Total error count:" ,  error_count)

```
