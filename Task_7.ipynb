# 1: Smart Light Controller

```python
class Light:
    def __init__(self,light_name):
        self.light_name = light_name

# Method to show light status
    def status_light(self,action):
        if action.lower() == "on":
            print(f"{self.light_name} is on")
        else:
            print(f"{self.light_name} is off")

# Sample INput
l = Light("Bedroom Light")
l.status_light('On')    # action = on
        
```

# 2: Employee ID Card System

# 3: Mobile Contact Record


```python
class Phone_contact:
    def __init__(self,name,number):
        self.name = name
        self.number = number

# Displays Contact saved
    def save_display(self):
        print(f"Contact Saved",f"\nName: {self.name.title()}",f"\nPhone NO.: {self.number}")

#INputs
person = Phone_contact('Durgesh',"7773738393")
person.save_display()

```

# 4: Product Price Tag Generator


```python
class Product:
    def __init__(self,product,price):
        self.product = product
        self.price = price

    def formatted_PriceTag(self):
        print(f"Product: {self.product.title()}",f"\nPrice: {self.price} RS")
# INputs
vastu = Product('jeaNs','15,000')
vastu.formatted_PriceTag()
```

#  5: Movie Rating Display System


```python
class Movie:
    def __init__(self,name,rating):
        self.name = name
        self.rating = rating

# Displays movie details
    def display_details(self):
        print(f'Movie Name: {self.name.title()}',f'\nIMDB Rating: {self.rating}/5')

# INputs
mv = Movie('shershah',4.9)
mv.display_details()
```

# 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self,name,address):
        self.name = name
        self.address = address

# Displays details of customer
    def details(self):
        print(f"DELIVERY DETAILS" , f"\nCustomer: {self.name.title()}",f"\nAddress: {self.address.title()}")

# INputs
Insan = Delivery('vishwajit suryawanshi','pune')
Insan.details()
```
