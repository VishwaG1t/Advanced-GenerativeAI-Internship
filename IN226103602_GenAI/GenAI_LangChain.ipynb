```python
from langchain_core.prompts import PromptTemplate

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template_1.format(topic="Artificial Intelligence"))
```

```python
from langchain_core.prompts import PromptTemplate

template_2 = PromptTemplate(
    input_variables=["topic", "level"],
    template="Explain {topic} for a {level} level student"
)

print(template_2.format(topic="Machine Learning", level="beginner"))
```

```python
from langchain_core.prompts import PromptTemplate

styles = ["simple", "technical", "real-world example"]

template_3 = PromptTemplate(
    input_variables=["topic", "style"],
    template="Explain {topic} in a {style} way"
)

for style in styles:
    print(template_3.format(topic="Deep Learning", style=style))
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert AI tutor"),
    ("user", "Explain {topic} clearly")
])

print(chat_template.format(topic="Neural Networks"))
```

```python
def validate(topic):
    if len(topic) < 3:
        return "Invalid input"
    return f"Explain {topic} in simple terms"

print(validate("AI"))
```
